````python
class Car:
    def __init__(self, brand, model):  # Constructor
        self.brand = brand
        self.model = model

    def drive(self):  # Method
        return f"{self.brand} {self.model} is driving!"

# Create objects
car1 = Car("Tesla", "Model S")
car2 = Car("BMW", "X5")

print(car1.drive())  # Tesla Model S is driving!
print(car2.drive())  # BMW X5 is driving!
````

````python
class House:
    def __init__(self, model, year, house_type): # Constructor
        self.model = model
        self.year = year
        self.house_type = house_type

    def built_home(self): # Method
        return f"House {self.house_type} and model {self.model} is built on {self.house_type}"

# Create Objects
house1 = House("Galactic", "1575", "space")
house2 = House("Milkyway", "0001", "Thirodentic")

print(house1.built_home())
print(house2.built_home())
````

## Inheritance

Allows a class (child) to reuse code from another class (parent).

````python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        return f"{self.brand} started!"

class Car(Vehicle):  # Car inherits from Vehicle
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

car = Car("Tesla", "Model 3")
print(car.start())  # Tesla started!
````

## Polymorphism

“Many forms” → Same method name but different behaviors.

````python
class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

# Polymorphism: same method, different output
animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())
````

## Encapsulation

Restricting direct access to data → using private/protected attributes.

````python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable (name mangling)

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())  # 1500
# print(account.__balance) ❌ Error (private)
````

## Abstraction

Hiding complex implementation details and exposing only the necessary interface.

````python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r

circle = Circle(5)
print(circle.area())  # 78.5
````

## Special Methods (Magic / Dunder methods)

Python classes come with special double-underscore methods (__method__) that give objects special behaviors.

🔹 __init__

Constructor → runs automatically when an object is created.

````python
class Person:
    def __init__(self, name):
        self.name = name

p = Person("Alice")
print(p.name)  # Alice
````

## __str__ vs __repr__

__str__ → Human-readable string (for users).

__repr__ → Developer/debug representation.

````python
class Person:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Person: {self.name}"   # For users

    def __repr__(self):
        return f"Person(name='{self.name}')"  # For developers

p = Person("Alice")
print(str(p))   # Person: Alice
print(repr(p))  # Person(name='Alice')
````

## __eq__ and __hash__

Used for object comparison and hashing (e.g., in sets, dict keys).

````python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __hash__(self):
        return hash((self.x, self.y))

p1 = Point(1, 2)
p2 = Point(1, 2)

print(p1 == p2)        # True (__eq__)
print({p1, p2})        # Only 1 element because __hash__ makes them equal
````

## __call__

Makes an object callable like a function.

````python
class Greeter:
    def __init__(self, name):
        self.name = name

    def __call__(self):
        return f"Hello, {self.name}!"

greet = Greeter("Rami")
print(greet())  # Hello, Rami!
````

# 🧑‍💻 Object-Oriented Programming (OOP) in Python

OOP is a programming paradigm where we model **real-world entities as objects** that have **attributes (data)** and **methods (behavior)**.

---

## 1️⃣ Class & Object

* **Class** → Blueprint/template.
* **Object** → Instance of a class.

```python
class Car:
    def __init__(self, brand, model):  # Constructor
        self.brand = brand
        self.model = model

    def drive(self):
        return f"{self.brand} {self.model} is driving."

# Object creation
car1 = Car("Tesla", "Model S")
print(car1.drive())  # Tesla Model S is driving.
```

---

## 2️⃣ Encapsulation

Wrapping **data + methods** together and restricting direct access.

* Use **private/protected** attributes to hide details.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())  # 1500
# print(account.__balance)  # ❌ AttributeError
```

---

## 3️⃣ Abstraction

Hiding **implementation details** and exposing only essentials using **abstract classes**.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r

circle = Circle(5)
print(circle.area())  # 78.5
```

---

## 4️⃣ Inheritance

A child class **inherits** properties/methods from a parent class.

### 🔹 Single Inheritance

```python
class Animal:
    def speak(self):
        return "I make a sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

dog = Dog()
print(dog.speak())  # Woof!
```

### 🔹 Multiple Inheritance

```python
class Father:
    def skills(self):
        return "Gardening"

class Mother:
    def skills(self):
        return "Cooking"

class Child(Father, Mother):
    pass

c = Child()
print(c.skills())  # Gardening (Father is first in MRO)
```

### 🔹 Multilevel Inheritance

```python
class Grandparent:
    def house(self):
        return "Big house"

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

c = Child()
print(c.house())  # Big house
```

### 🔹 Hierarchical Inheritance

```python
class Parent:
    def work(self):
        return "Parent working"

class Child1(Parent):
    pass

class Child2(Parent):
    pass

print(Child1().work())  # Parent working
print(Child2().work())  # Parent working
```

### 🔹 Hybrid Inheritance

(Combination of multiple inheritance types)

```python
class A:
    def show(self): return "A"

class B(A): pass
class C(A): pass
class D(B, C): pass  # Hybrid

print(D().show())  # A
```

---

## 5️⃣ Polymorphism

“Many forms” → same method name, different behaviors.

```python
class Bird:
    def sound(self): return "Chirp"

class Dog:
    def sound(self): return "Woof"

# Polymorphism
for animal in [Bird(), Dog()]:
    print(animal.sound())
```

---

## 6️⃣ Method Overloading (⚠️ Python doesn’t support it directly)

* **In Java/C++** → same method name, different parameters.
* **In Python** → handled with default arguments or `*args`.

```python
class MathOps:
    def add(self, a, b=0, c=0):
        return a + b + c

obj = MathOps()
print(obj.add(5))        # 5
print(obj.add(5, 10))    # 15
print(obj.add(5, 10, 20))# 35
```

---

## 7️⃣ Method Overriding

Child class **redefines** parent class method.

```python
class Parent:
    def show(self):
        return "Parent class method"

class Child(Parent):
    def show(self):
        return "Child class method"

c = Child()
print(c.show())  # Child class method
```

---

# ✅ Quick Summary

* **Class/Object** → Blueprint & instance
* **Encapsulation** → Hide data (private/protected attributes)
* **Abstraction** → Hiding implementation via abstract classes
* **Inheritance** → Reusing code (Single, Multiple, Multilevel, Hierarchical, Hybrid)
* **Polymorphism** → Same interface, different implementations
* **Method Overloading** → Same method, multiple params (simulated in Python)
* **Method Overriding** → Redefining parent’s method in child
