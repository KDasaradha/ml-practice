### 🔄 Update Poetry

```powershell
poetry self update
```

### 🔍 Verify

```powershell
poetry --version
```

### 🆕 New Project

```powershell
poetry new my_project
```

```
my_project/
├── pyproject.toml
├── README.rst
├── my_project/
│   └── __init__.py
└── tests/
    └── __init__.py
```

### 🏗️ Initialize in Existing Project

```powershell
cd existing_project
poetry init
```

* Interactive prompt will ask for dependencies.
* Generates a **`pyproject.toml`** file.

## 3. Starting a Project with uv

### 3.1 New Project (scaffolding)

To create a new uv-managed project:

```bash
uv init my_project
```

This does the following:

* Creates a directory `my_project` (if not already existing). ([Real Python][2])
* Initializes a Git repository (by default). ([Real Python][2])
* Creates essential files such as:

  * `pyproject.toml`
  * `.python-version` (records the default Python version for the project) ([Real Python][2])
  * `README.md`
  * A sample Python file (e.g. `hello.py`) ([Real Python][2])
  * `.gitignore` ([Real Python][2])

After that, `cd my_project` to enter your project.

---

### 3.2 Adopt uv in an Existing Project

If you have an existing project (with a `pyproject.toml` or without), you can do:

```bash

```

* If a `pyproject.toml` exists, uv tries not to overwrite it. ([Real Python][2])
* It will scaffold missing project config files, but won’t override your existing Python files. ([Real Python][2])

In short: it helps bring uv’s structure into your existing project without destroying what you already have. ([Real Python][2])

## 9. Differences vs Other Tools (Poetry / pip / venv)

* **Speed**: uv is substantially faster than pip, virtualenv, and even pip-tools in many cases (due to parallelism & optimized resolver) ([DataCamp][6])
* **Unified Tool**: Combines what pip, virtualenv, pipx, poetry, pyenv, etc. do into one CLI. ([GitHub][1])
* **Lockfile + reproducibility**: `uv.lock` ensures consistent dependencies across machines. ([Real Python][2])
* **Inline script dependencies**: Unique feature: scripts can declare their own dependencies within the file. ([GitHub][1])
* **Python version management**: uv can install different Python versions for your project. Poetry typically delegates to a system Python or external tools. ([Astral Docs][3])
* **Project scaffolding**: uv provides `uv init`, similar to Poetry’s `poetry new` / `poetry init`. ([GitHub][1])

## 10. Tips, Caveats & Best Practices

* Always commit `uv.lock` to version control; it ensures reproducibility. Avoid manually editing it.
* If dependencies change (e.g. manually editing `pyproject.toml`), run `uv lock` and `uv sync`.
* Use `uv run` as your primary command instead of manually invoking Python. It handles environment setup automatically.
* When needing to install into system Python (rare), use `--system` carefully, especially in CI. ([Stack Overflow][5])
* For large multi-module projects, consider using uv workspaces to share dependencies.
* Keep uv up to date (via `uv self update` if you installed via standalone).
* Use `uv tool install` or `uvx` for CLI tools (e.g. linters, formatters) rather than mixing pip or pipx manually.
* If you’re migrating from pip/requirements or Poetry, you can often adopt uv gradually (installing dependencies via `uv pip install -r requirements.txt`) and then switch to uv’s project commands. ([Astral Docs][4])

## 🔹 What is a **Makefile**?

* A **Makefile** is a special file used by the `make` build automation tool.
* It defines a set of **rules** (targets, dependencies, and commands) to automate repetitive tasks.
* Originally popular in **C/C++ projects**, but also widely used in **Python projects** to simplify workflows.

Think of it as a **task runner** — similar to `npm scripts` in Node.js, but system-wide.

---

## 🔹 Why use Makefile in Python projects?

Python projects often need commands like:

* Setting up virtual environments
* Installing dependencies
* Running tests
* Linting/formatting code
* Building docs
* Packaging & deployment

Instead of remembering long commands, you just type `make <task>`.

---

## 🔹 Example Use Case in Python

### 1. Without Makefile

You might run commands like:

```bash
python -m venv .venv
source .venv/bin/activate
pip install -r requirements.txt
pytest -v
black .
```

### 2. With Makefile

You can define those tasks in `Makefile`:

```makefile
# Makefile for Python Project

.PHONY: install test format run clean

# Create virtual environment and install dependencies
install:
	python -m venv .venv
	. .venv/bin/activate && pip install -r requirements.txt

# Run tests
test:
	pytest -v

# Format code
format:
	black .

# Run the application
run:
	. .venv/bin/activate && python main.py

# Clean up caches and venv
clean:
	rm -rf __pycache__ .pytest_cache .venv
```

Now you just run:

```bash
make install
make test
make format
make run
make clean
```

---

## 🔹 Benefits

✅ **Consistency** — teammates run the same commands
✅ **Convenience** — shorter, easier-to-remember commands
✅ **Cross-project reuse** — you can copy the Makefile across projects
✅ **Integration** — works with CI/CD pipelines

---

## 🔹 Where it fits in a Python project

Typical project structure:

```
my_project/
│── pyproject.toml     # Poetry or other package manager config
│── Makefile           # Automates tasks
│── requirements.txt   # Optional fallback
│── src/
│    └── main.py
│── tests/
│    └── test_main.py
```

---

⚡ In modern Python projects, a **Makefile** is often used alongside **Poetry**, **uv**, or **pip-tools** — it just acts as a **wrapper for common commands** so you don’t have to memorize them.

# 🛠️ Makefile Cheat Sheet for Python Projects

Makefile lets you automate tasks (setup, testing, linting, running) in your Python projects.
Think of it as a **task runner** like `npm scripts` in JavaScript.

---

## 📌 1. Installation


## 📌 2. Create a Makefile

Create a file named `Makefile` (no extension) in your project root:

```bash
touch Makefile
```

---

## 📌 3. Basic Syntax

```makefile
target: dependencies
	command1
	command2
```

* **target** → task name (`make target`)
* **dependencies** → other tasks that must run before
* **commands** → shell commands (must start with a tab)

Example:

```makefile
hello:
	echo "Hello, World!"
```

Run with:

```bash
make hello
```

---

## 📌 4. Common Python Project Tasks

### 🎯 Virtual Environment + Install

```makefile
.PHONY: install

install:
	python -m venv .venv
	. .venv/bin/activate && pip install -r requirements.txt
```

### 🎯 Run Tests

```makefile
.PHONY: test

test:
	pytest -v
```

### 🎯 Format + Lint

```makefile
.PHONY: format lint
```
