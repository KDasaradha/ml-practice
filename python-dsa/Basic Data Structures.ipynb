## 🧱 **Lists in Python (Dynamic Arrays)**

Python **lists** are *dynamic arrays*, meaning they can change size at runtime — you can add, remove, or modify elements easily.

---

### 🔹 1. Dynamic Arrays

In languages like C/C++, arrays have fixed sizes.
But in Python, lists automatically resize when needed.

#### ✅ Example:

```python
nums = [1, 2, 3]
print(nums)

nums.append(4)   # Add new element
nums.append(5)
print(nums)
```

🧠 **Behind the scenes:**

* Python over-allocates memory to reduce the cost of resizing.
* When the list’s capacity is exceeded, it creates a **new larger array**, copies elements over, and frees the old one.

This makes:

* `append()` → **Amortized O(1)**
* but occasional **O(n)** when resizing happens.

#### 💡 Memory efficiency:

You can check memory size using:

```python
import sys
nums = []
for i in range(10):
    nums.append(i)
    print(i, sys.getsizeof(nums))
```

You’ll notice the list size doesn’t increase every single time — because of *over-allocation*.

---

### 🔹 2. Indexing & Slicing

#### ✅ Indexing

You can access elements using their position (starting at index `0`):

```python
nums = [10, 20, 30, 40, 50]
print(nums[0])   # 10
print(nums[-1])  # 50 (last element)
```

#### ✅ Slicing

You can extract parts of a list easily:

```python
nums = [10, 20, 30, 40, 50]
print(nums[1:4])    # [20, 30, 40]
print(nums[:3])     # [10, 20, 30]
print(nums[::2])    # [10, 30, 50] (step of 2)
```

🧠 *Slicing returns a new list — not a reference!*

---

### 🔹 3. Shallow vs Deep Copy

When copying lists, you must be careful with *references*.

#### ⚠️ Example:

```python
a = [[1, 2], [3, 4]]
b = a  # same reference
b[0][0] = 99
print(a)  # [[99, 2], [3, 4]]
```

Both changed — because they point to the same memory.

#### ✅ Shallow Copy

Copies only the outer list:

```python
import copy
a = [[1, 2], [3, 4]]
b = a.copy()  # or list(a)
b[0][0] = 99
print(a)  # [[99, 2], [3, 4]]
```

#### ✅ Deep Copy

Copies *everything recursively*:

```python
import copy
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)
b[0][0] = 99
print(a)  # [[1, 2], [3, 4]]
```

🧠 **Rule:**

* `copy()` → only copies top-level
* `deepcopy()` → full recursive copy

## 🧩 **Tuples in Python**

A **tuple** is an *immutable*, *ordered* collection of elements.
Think of it as a “read-only list” — once created, it cannot be modified.

---

### 🔹 1. Immutability

#### ✅ Definition:

```python
numbers = (10, 20, 30)
print(numbers[0])  # 10
```

Tuples **cannot** be:

* Modified
* Appended to
* Removed from

Example:

```python
numbers = (1, 2, 3)
numbers[0] = 99     # ❌ TypeError
numbers.append(4)   # ❌ AttributeError
```

So tuples are **immutable**, meaning:

* The contents cannot be changed after creation.
* But if the tuple contains mutable objects (like a list), those can still be changed inside.

#### ⚠️ Example of mutable inside immutable:

```python
t = (1, [2, 3], 4)
t[1][0] = 99
print(t)  # (1, [99, 3], 4)
```

→ Tuple itself is immutable, but elements *inside it* can be changed if they are mutable.

---

### 🔹 2. Tuple Creation

```python
t1 = (1, 2, 3)
t2 = 1, 2, 3  # parentheses optional
t3 = (1,)     # single element tuple → comma is required!
t4 = tuple([1, 2, 3])
```

If you forget the comma:

```python
x = (1)
print(type(x))  # int ❌
x = (1,)
print(type(x))  # tuple ✅
```

---

### 🔹 3. Tuple Unpacking

You can directly “unpack” tuples into variables.

#### ✅ Basic Example:

```python
person = ("Alice", 25, "Engineer")
name, age, job = person
print(name, age, job)
```

#### ✅ Ignoring values:

```python
data = (1, 2, 3, 4)
a, _, _, d = data
print(a, d)  # 1 4
```

#### ✅ Unpacking with `*` (extended unpacking):

```python
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(first)   # 1
print(middle)  # [2, 3, 4]
print(last)    # 5
```

---

### 🔹 4. Why Tuples?

✅ **Performance** — faster than lists (stored in contiguous memory).
✅ **Safety** — can be used as *keys in dictionaries* (since immutable).
✅ **Memory efficient** — uses less memory than lists.

```python
import sys
lst = [1, 2, 3, 4, 5]
tup = (1, 2, 3, 4, 5)

print(sys.getsizeof(lst))  # e.g. 104
print(sys.getsizeof(tup))  # e.g. 88 (smaller)
```

### 🔹 5. NamedTuple / Dataclasses

Both provide **structure** and **readability** to tuples.

---

#### 🏷️ `collections.namedtuple`

A **NamedTuple** is like a lightweight class —
you can access fields by *name* instead of *index*.

```python
from collections import namedtuple

Person = namedtuple("Person", ["name", "age", "job"])
p = Person("Alice", 25, "Engineer")

print(p.name)  # Alice
print(p.age)   # 25
print(p[0])    # Alice (still works like tuple)
```

✅ Immutable
✅ Memory-efficient
✅ Ideal for lightweight record storage

---

#### 🧰 `dataclasses.dataclass`

Dataclasses are more flexible and class-like.
They can be **mutable or immutable** depending on configuration.

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class Person:
    name: str
    age: int
    job: str

p = Person("Alice", 25, "Engineer")
print(p.name)
# p.age = 30  # ❌ FrozenInstanceError (immutable)
```

✅ Easier to use
✅ Supports defaults, type hints, comparisons, etc.
✅ Use `frozen=True` to make it immutable like a NamedTuple

### 🔹 5. Hashing Concept (Under the Hood)

Now, let’s understand **how sets actually work** 🧠

#### ⚙️ Internal Mechanism:

* Python sets are implemented using **hash tables** (like dictionaries).
* Each element in a set is stored based on its **hash value** (computed using `hash()` function).
* Hashing converts an object into an integer (the “hash”), which determines where it’s stored in memory.

#### ✅ Example:

```python
print(hash("apple"))
print(hash(42))
print(hash((1, 2, 3)))
```

* Each element is placed in a “bucket” based on its hash.
* This allows **O(1)** average-time complexity for:

  * `add()`
  * `remove()`
  * `in` checks (membership)

#### ⚠️ Note:

Mutable types (like lists or dicts) are *not hashable* → their content can change, which would break the hash consistency.

That’s why:

```python
s = set()
s.add([1, 2])  # ❌ TypeError
```

But this works:

```python
s = set()
s.add((1, 2))  # ✅ tuples are hashable
```

## ⚙️ How Hashing Works (Under the Hood)

---

### 🔹 1. The Concept of Hashing

**Hashing** is the process of converting an object (like a string or number) into a **fixed-size integer value** called a *hash code*.

This hash code is then used as an **index** to store the element inside a hash table (an internal array).

#### Example:

```python
hash("apple")  # → e.g. 3565435892458
hash("banana") # → e.g. -23442457363
```

Each object → unique(ish) integer
Stored in memory based on that hash.

---

### 🔹 2. Hash Table Structure

A **hash table** is essentially an **array of buckets (slots)**.
Each bucket can store one or more values, based on their hash.

```
Index:   0     1     2     3     4     5
Data:   ---  "cat"  ---  "dog"  ---  "bat"
```

When you do:

```python
s = {"cat", "dog", "bat"}
```

Python internally:

1. Computes each element’s hash.
2. Determines where to place it → `index = hash(element) % table_size`.
3. Stores it there.

---

### 🔹 3. Hash Collisions

A **collision** happens when two different elements have the same hash index.

Example:

```python
hash("dog") % 8 == hash("god") % 8  # could happen
```

Python handles this using **open addressing** (specifically *probing*):

* If a slot is already filled, it looks for the *next available* slot.
* This process continues until it finds an empty space.

This ensures every element still gets stored, even with collisions.

---

### 🔹 4. Rehashing (Resizing)

When too many elements are added:

* The load factor (ratio of elements to table size) increases.
* Lookup time slows down.

So Python **automatically resizes (rehashes)** the table:

* Creates a new, larger array.
* Recomputes hash positions for all existing elements.

This keeps the average operation time **O(1)**.

---

### 🔹 5. Why Sets are Fast (Big-O Analysis)

| Operation  | Average Time | Worst Case | Explanation               |
| ---------- | ------------ | ---------- | ------------------------- |
| `add()`    | O(1)         | O(n)       | Insert via hash           |
| `remove()` | O(1)         | O(n)       | Find via hash             |
| `x in set` | O(1)         | O(n)       | Membership check via hash |
| `len(set)` | O(1)         | O(1)       | Stored count              |

Even though worst case is O(n) (e.g., many collisions),
Python’s hash functions and table resizing keep it practically O(1).

---

### 🔹 6. Hashable vs Unhashable

An object can be stored in a set **only if it’s hashable**:

* Hashable means:

  * It has a fixed hash value (`__hash__()` method)
  * It’s immutable (`__eq__()` never changes outcome)

✅ Hashable:

* int, float, str, tuple (of hashables), frozenset

❌ Unhashable:

* list, dict, set (because they can change)

---

### 🔹 7. frozenset (Immutable Set)

If you need a **set that cannot be modified** and can itself be used as a dictionary key or stored in another set → use **frozenset**.

```python
a = frozenset([1, 2, 3])
b = frozenset([3, 4, 5])

print(a | b)  # Union → frozenset({1, 2, 3, 4, 5})
# a.add(6)  ❌ AttributeError (immutable)
```

✅ Hashable
✅ Supports all set operations
✅ Safer when you need constant data

---

### 🔹 8. Visual Example — Hash Table View

Let’s visualize adding elements to a set:

```python
s = set()
s.add("apple")   # hash('apple') → index 4
s.add("banana")  # hash('banana') → index 7
s.add("grape")   # hash('grape') → index 4 (collision!)
```

After internal probing:

```
[ 0 ][ 1 ][ 2 ][ 3 ][4:"apple"][5:"grape"][6][7:"banana"]
```

Even though `"grape"` had the same hash index as `"apple"`,
Python probed to the next available slot.

---

### 🔹 9. Practical Hash Example (to understand collisions)

```python
def simple_hash(s):
    return sum(ord(c) for c in s) % 10

words = ["bat", "tab", "cat"]
for w in words:
    print(w, "→", simple_hash(w))
```

🧠 Notice how `"bat"` and `"tab"` may have the same hash → collision!

### 🔹 10. Summary — How Sets Use Hashing

| -------------------- | --------------------------------------------- |

| **Collision**        | Handled using open addressing (probing)       |
| **Rehashing**        | Resize table when load factor grows           |
| **Hashable Objects** | Immutable types with fixed hash value         |
| **frozenset**        | Immutable, hashable version of a set          |

## 🗄️ **Dictionaries in Python**

A **dictionary** is an **unordered collection of key-value pairs**. Keys must be **hashable**, values can be anything.

---

### 🔹 1. Hash Table Implementation (Under the Hood)

Dictionaries are **hash tables** like sets:

1. Each key is **hashed** → hash value determines the index (bucket) in an internal array.
2. Values are stored alongside keys.
3. Collisions are resolved using **open addressing** (probing) or Python’s internal optimizations.
4. Lookup, insertion, and deletion are **O(1) average**.

#### Example of internal idea:

```python
d = {"apple": 10, "banana": 20}
# hash("apple") → index 3
# hash("banana") → index 7
# Value stored in that bucket
```

---

### 🔹 2. Key-Value Operations

#### ✅ Create dictionary

```python
# Using braces
d = {"a": 1, "b": 2, "c": 3}

# Using dict() constructor
d = dict(x=1, y=2)
```

#### ✅ Access / Modify

```python
d = {"name": "Alice", "age": 25}
print(d["name"])  # Alice
d["age"] = 26     # modify
d["city"] = "NY"  # add new key
```

#### ✅ Delete

```python
del d["age"]      # removes key
d.pop("city")     # removes and returns value
d.popitem()       # removes last inserted item
```

#### ✅ Membership

```python
"name" in d       # True
25 in d           # False (checks keys only)
```

#### ✅ Iteration

```python
for key in d:
    print(key, d[key])

for key, value in d.items():
    print(key, value)
```

---

### 🔹 3. Dictionary Comprehensions

Like list comprehensions, you can create dicts dynamically:

```python
# Square numbers
squares = {x: x*x for x in range(5)}
print(squares)  # {0:0, 1:1, 2:4, 3:9, 4:16}

# Conditional comprehension
even_squares = {x: x*x for x in range(10) if x%2==0}
print(even_squares)  # {0:0, 2:4, 4:16, 6:36, 8:64}
```

✅ Fast, readable, and efficient.

---

### 🔹 4. Specialized Dictionaries

Python’s `collections` module provides enhanced dictionary types.

#### 4.1 `defaultdict`

* Provides default value for missing keys to avoid `KeyError`.

```python
from collections import defaultdict

dd = defaultdict(int)  # default 0
dd["a"] += 1
dd["b"] += 2
print(dd)  # defaultdict(<class 'int'>, {'a': 1, 'b': 2})
```

You can also use `list`, `set`, etc.

```python
dd = defaultdict(list)
dd["fruits"].append("apple")
print(dd)  # {'fruits': ['apple']}
```

---

#### 4.2 `Counter`

* Specialized dict for **counting hashable objects**.

```python
from collections import Counter

words = ["apple", "banana", "apple", "orange", "banana", "apple"]
c = Counter(words)
print(c)  # Counter({'apple': 3, 'banana': 2, 'orange': 1})

print(c.most_common(2))  # [('apple', 3), ('banana', 2)]
```

✅ Very useful in frequency-based problems.

---

#### 4.3 `OrderedDict`

* Keeps insertion order (Python ≥3.7 regular dict already preserves order).

```python
from collections import OrderedDict

od = OrderedDict()
od["a"] = 1
od["b"] = 2
od["c"] = 3

for k, v in od.items():
    print(k, v)  # a 1, b 2, c 3
```

* Supports operations like `move_to_end(key)`.

> Like sets, dictionary keys must be **hashable** (immutable types like int, str, tuple).

### 🔹 6. Practical Examples

#### Count letters in a string using `Counter`:

```python
from collections import Counter

s = "abracadabra"
letter_count = Counter(s)
print(letter_count)  # Counter({'a':5, 'b':2, 'r':2, 'c':1, 'd':1})
```

#### Group items by category using `defaultdict`:

```python
items = [("fruit", "apple"), ("fruit", "banana"), ("veg", "carrot")]
grouped = defaultdict(list)
for category, name in items:
    grouped[category].append(name)

print(grouped)  
# defaultdict(<class 'list'>, {'fruit': ['apple', 'banana'], 'veg': ['carrot']})
```

#### Dictionary comprehension:

```python
squares = {x: x*x for x in range(1,6)}
print(squares)  # {1:1, 2:4, 3:9, 4:16, 5:25}
```
