## **1. Stack Overview**

A **stack** is a **linear data structure** that follows the **LIFO (Last In, First Out)** principle.

* **LIFO** means the last element added is the first one to be removed.
* Think of a **stack of plates**: you put new plates on top and remove plates from the top.

**Basic operations:**

| Operation           | Description                              |
| ------------------- | ---------------------------------------- |
| `push(x)`           | Add element `x` to the top of the stack  |
| `pop()`             | Remove and return the top element        |
| `peek()` or `top()` | View the top element without removing it |
| `is_empty()`        | Check if the stack is empty              |

---

## **2. Implementation in Python**

### **a) Using a List**

```python
# Stack implementation using Python list
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None  # or raise Exception

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


# Example usage
stack = Stack()
stack.push(10)
stack.push(20)
print(stack.pop())   # 20
print(stack.peek())  # 10
print(stack.is_empty())  # False
```

---

### **b) Using `collections.deque`**

`deque` is faster for stack operations than lists for large data because it avoids resizing overhead.

```python
from collections import deque

stack = deque()
stack.append(10)  # push
stack.append(20)
print(stack.pop())  # 20
print(stack[-1])    # peek: 10
```

### **b) Balanced Parentheses**

* Check if an expression has matching `()`, `{}`, `[]`.
* Algorithm:

  1. Traverse the string.
  2. Push opening brackets to stack.
  3. For closing brackets, check if top of stack matches. If yes, pop; otherwise, invalid.
  4. At the end, stack should be empty.

```python
def is_balanced(expression):
    stack = []
    bracket_map = {')': '(', ']': '[', '}': '{'}
    
    for char in expression:
        if char in "([{":
            stack.append(char)
        elif char in ")]}":
            if not stack or stack[-1] != bracket_map[char]:
                return False
            stack.pop()
    return not stack

# Test
print(is_balanced("{[()]}"))  # True
print(is_balanced("{[(])}"))  # False
```

## **Backtracking Example: Maze Solver (Iterative with Stack)**

**Problem:**
Given a 2D maze (`0 = path, 1 = wall`), find a path from the start `(0,0)` to the end `(n-1,m-1)`.

**Algorithm:**

1. Push the starting position to the stack.
2. While the stack is not empty:

   * Pop the top position.
   * If it's the goal, return the path.
   * Otherwise, push all valid neighboring positions (up, down, left, right) that are not visited.
3. Mark visited cells to avoid cycles.

---

### **Python Implementation**

```python
def solve_maze(maze):
    rows, cols = len(maze), len(maze[0])
    stack = [((0, 0), [(0, 0)])]  # (current_position, path_taken)
    visited = set()

    while stack:
        (x, y), path = stack.pop()
        
        if (x, y) in visited:
            continue
        visited.add((x, y))
        
        # Goal reached
        if (x, y) == (rows - 1, cols - 1):
            return path
        
        # Move directions: down, right, up, left
        for dx, dy in [(1,0),(0,1),(-1,0),(0,-1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < rows and 0 <= ny < cols and maze[nx][ny] == 0:
                stack.append(((nx, ny), path + [(nx, ny)]))
    
    return None  # No path found

# Example maze (0 = path, 1 = wall)
maze = [
    [0, 1, 0, 0],
    [0, 0, 0, 1],
    [1, 0, 1, 0],
    [0, 0, 0, 0]
]

path = solve_maze(maze)
print("Path found:" if path else "No path!", path)
```

---

### **Explanation**

* Stack keeps the **last explored path** on top.
* If a dead-end is reached, the algorithm **pops** and backtracks to the previous choice.
* No recursion is used; the stack handles all backtracking logic.
* This is exactly how stacks help in **backtracking problems** like **Sudoku, N-Queens, and permutations**.

* Stack = LIFO → last choice is explored first.
* Backtracking is naturally implemented using stacks (iterative) or recursion (which uses call stack internally).

## **1. Expression Types**

1. **Infix:** Operator between operands → `A + B`
2. **Postfix (Reverse Polish Notation):** Operator after operands → `A B +`
3. **Prefix:** Operator before operands → `+ A B`

We usually convert **infix → postfix** because postfix can be **easily evaluated using a stack** without worrying about parentheses or operator precedence.

---

## **2. Infix → Postfix Conversion**

**Rules using Stack:**

1. Operands → add directly to output.
2. Opening bracket `(` → push to stack.
3. Closing bracket `)` → pop from stack until `(` is found.
4. Operators → pop from stack to output while the top has **higher or equal precedence**, then push the current operator.

**Operator Precedence (Higher number = higher precedence):**

```python
precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
```

---

### **Python Implementation**

```python
def infix_to_postfix(expression):
    stack = []
    output = []
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}

    for char in expression:
        if char.isalnum():  # Operand
            output.append(char)
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()  # remove '('
        else:  # Operator
            while stack and stack[-1] != '(' and precedence[stack[-1]] >= precedence[char]:
                output.append(stack.pop())
            stack.append(char)

    while stack:
        output.append(stack.pop())

    return ''.join(output)

# Example
expr = "A+B*(C-D)"
postfix = infix_to_postfix(expr)
print("Postfix:", postfix)  # Output: ABCD-*+
```

## **3. Postfix Expression Evaluation**

**Rules using Stack:**

1. Read the postfix expression from left to right.
2. Operand → push onto stack.
3. Operator → pop **two operands** from stack, apply operator, push result back.
4. At the end, stack contains the result.

---

### **Python Implementation (Numeric)**

```python
def evaluate_postfix(expression):
    stack = []
    
    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:  # Operator
            b = stack.pop()
            a = stack.pop()
            if char == '+':
                stack.append(a + b)
            elif char == '-':
                stack.append(a - b)
            elif char == '*':
                stack.append(a * b)
            elif char == '/':
                stack.append(a // b)  # integer division
            elif char == '^':
                stack.append(a ** b)
    
    return stack[0]

# Example
postfix_expr = "23*5+"
result = evaluate_postfix(postfix_expr)
print("Result:", result)  # Output: 11 (2*3 +5)
```

## **Complete Infix Expression Evaluator**

```python
def infix_to_postfix(expression):
    stack = []
    output = []
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    i = 0
    while i < len(expression):
        char = expression[i]

        # Multi-digit number handling
        if char.isdigit():
            num = char
            while i + 1 < len(expression) and expression[i + 1].isdigit():
                i += 1
                num += expression[i]
            output.append(num)
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()  # remove '('
        else:  # Operator
            while stack and stack[-1] != '(' and precedence.get(stack[-1], 0) >= precedence[char]:
                output.append(stack.pop())
            stack.append(char)
        i += 1

    while stack:
        output.append(stack.pop())

    return output  # list of tokens

def evaluate_postfix(postfix):
    stack = []
    for token in postfix:
        if token.isdigit():
            stack.append(int(token))
        else:
            b = stack.pop()
            a = stack.pop()
            if token == '+':
                stack.append(a + b)
            elif token == '-':
                stack.append(a - b)
            elif token == '*':
                stack.append(a * b)
            elif token == '/':
                stack.append(a // b)  # integer division
            elif token == '^':
                stack.append(a ** b)
    return stack[0]

# Example
expression = "12+(23*3)-(4^2)"
postfix = infix_to_postfix(expression)
print("Postfix:", postfix)  # ['12', '23', '3', '*', '+', '4', '2', '^', '-']
result = evaluate_postfix(postfix)
print("Result:", result)    # 12 + (23*3) - (4^2) = 71
```

2. **Stack handles operators**: Ensures correct precedence and parentheses handling.
3. **Postfix evaluation**: Pops last two operands, applies operator, pushes result back.
4. Works for **any valid infix arithmetic expression** with integers.

## **1. Queue Overview**

A **queue** is a linear data structure that follows the **FIFO (First In, First Out)** principle:

* **First** element added → **First** element removed.
* Think of a queue in real life, like a line at a ticket counter.

**Basic operations:**

1. **enqueue** → Add an element at the rear.
2. **dequeue** → Remove an element from the front.
3. **peek/front** → Look at the front element without removing it.
4. **isEmpty** → Check if the queue is empty.
5. **size** → Number of elements in the queue.

---

## **2. Variants of Queues**

### **a. Simple Queue**

* Standard FIFO queue.
* Elements are added at the rear and removed from the front.

```python
from collections import deque

queue = deque()

# Enqueue
queue.append(1)
queue.append(2)
queue.append(3)

# Dequeue
print(queue.popleft())  # 1
print(queue.popleft())  # 2
```

---

### **b. Circular Queue**

* Overcomes the limitation of fixed-size arrays where space cannot be reused.
* Rear and front wrap around to the start when the end is reached.

**Visual:**

```
[ _ _ _ _ ]   front=0, rear=0
enqueue 1 -> [ 1 _ _ _ ] front=0, rear=1
enqueue 2 -> [ 1 2 _ _ ] front=0, rear=2
dequeue   -> [ _ 2 _ _ ] front=1, rear=2
```

```python
class CircularQueue:
    def __init__(self, k):
        self.queue = [None] * k
        self.front = self.rear = -1
        self.size = k

    def enqueue(self, val):
        if ((self.rear + 1) % self.size == self.front):
            print("Queue is full")
            return
        if self.front == -1:
            self.front = 0
        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = val

    def dequeue(self):
        if self.front == -1:
            print("Queue is empty")
            return
        result = self.queue[self.front]
        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.size
        return result
```

---

### **c. Priority Queue**

* Each element has a **priority**.
* Element with **highest priority** is dequeued first.
* Can be implemented using `heapq` in Python (min-heap by default).

```python
import heapq

pq = []
heapq.heappush(pq, (2, "Task 2"))  # (priority, value)
heapq.heappush(pq, (1, "Task 1"))
heapq.heappush(pq, (3, "Task 3"))

while pq:
    print(heapq.heappop(pq)[1])  # Task 1, Task 2, Task 3
```

---

### **d. Deque (Double-Ended Queue)**

* Can add/remove elements from **both ends**.
* Implemented via `collections.deque`.

```python
from collections import deque

dq = deque([1, 2, 3])
dq.appendleft(0)  # Add to front
dq.append(4)      # Add to rear
print(dq)         # deque([0, 1, 2, 3, 4])

dq.popleft()      # Remove from front -> 0
dq.pop()          # Remove from rear -> 4
```

```python
print("\n=== Circular Queue ===")

class CircularQueue:
    def __init__(self, k):
        self.queue = [None] * k
        self.front = self.rear = -1
        self.size = k

    def enqueue(self, val):
        if (self.rear + 1) % self.size == self.front:
            print("Queue is full")
            return
        if self.front == -1:
            self.front = 0
        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = val

    def dequeue(self):
        if self.front == -1:
            print("Queue is empty")
            return
        result = self.queue[self.front]
        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.size
        return result

cq = CircularQueue(3)
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)  # Queue is full now
print("Dequeued:", cq.dequeue())
cq.enqueue(3)  # Space reused
print("Dequeued:", cq.dequeue())
```

## **1. CPU Scheduling (Using Queues)**

### **Concept**

* Operating systems use **queues** to schedule tasks.
* **Round Robin Scheduling** (time-sharing) uses a **simple queue**:

  1. Tasks are added to the queue.
  2. CPU executes each task for a fixed **time slice**.
  3. If task isn’t finished, it goes to the **rear of the queue**.

**Python Example: Round Robin Scheduling**

```python
from collections import deque

print("=== Round Robin CPU Scheduling ===")

tasks = deque([
    {"name": "Task1", "burst": 5},
    {"name": "Task2", "burst": 3},
    {"name": "Task3", "burst": 8},
])

time_slice = 2

while tasks:
    task = tasks.popleft()
    print(f"Executing {task['name']} for {time_slice} units")
    task['burst'] -= time_slice
    if task['burst'] > 0:
        tasks.append(task)  # Task not finished, go to rear
    else:
        print(f"{task['name']} completed")
```

**Output Example:**

```
Executing Task1 for 2 units
Executing Task2 for 2 units
Executing Task3 for 2 units
Executing Task1 for 2 units
Executing Task2 for 2 units
...
Task2 completed
...
```

✅ **Key Idea:** Queue ensures **fairness** — first come, first served.

## **2. Caching (FIFO & LRU)**

### **Concept**

* **Cache** stores frequently used data for faster access.
* **FIFO Cache:** First added item is removed when cache is full.
* **LRU (Least Recently Used) Cache:** Oldest accessed item is removed.

**Python Example: FIFO Cache using deque**

```python
from collections import deque

print("\n=== FIFO Cache Example ===")
cache_size = 3
cache = deque(maxlen=cache_size)

def access_page(page):
    if page in cache:
        print(f"Page {page} hit")
    else:
        print(f"Page {page} miss")
        cache.append(page)  # automatically removes oldest if full
    print("Cache:", list(cache))

pages = [1, 2, 3, 4, 2, 5]
for page in pages:
    access_page(page)
```

**Output:**

```
Page 1 miss
Cache: [1]
Page 2 miss
Cache: [1, 2]
Page 3 miss
Cache: [1, 2, 3]
Page 4 miss
Cache: [2, 3, 4]
Page 2 hit
Cache: [2, 3, 4]
Page 5 miss
Cache: [3, 4, 5]
```

**Python Example: LRU Cache using `OrderedDict`**

```python
from collections import OrderedDict

print("\n=== LRU Cache Example ===")
cache_size = 3
lru_cache = OrderedDict()

def access_page_lru(page):
    if page in lru_cache:
        print(f"Page {page} hit")
        lru_cache.move_to_end(page)  # Most recently used goes to end
    else:
        print(f"Page {page} miss")
        if len(lru_cache) >= cache_size:
            removed = lru_cache.popitem(last=False)  # Remove least recently used
            print(f"Removed {removed[0]}")
        lru_cache[page] = True
    print("Cache:", list(lru_cache.keys()))

pages = [1, 2, 3, 4, 2, 5]
for page in pages:
    access_page_lru(page)
```

**Output:**

```
Page 1 miss
Cache: [1]
Page 2 miss
Cache: [1, 2]
Page 3 miss
Cache: [1, 2, 3]
Page 4 miss
Removed 1
Cache: [2, 3, 4]
Page 2 hit
Cache: [3, 4, 2]
Page 5 miss
Removed 3
Cache: [4, 2, 5]
```

✅ **Key Idea:** Queues efficiently manage **FIFO or LRU cache eviction policies**.

## **2. Queue Variants**

### **a. Circular Queue**

* Reuses empty space after dequeue
* Useful for fixed-size queues

```python
class CircularQueue:
    def __init__(self, k):
        self.queue = [None]*k
        self.front = self.rear = -1
        self.size = k

    def enqueue(self, val):
        if (self.rear+1) % self.size == self.front:
            print("Full"); return
        if self.front == -1: self.front = 0
        self.rear = (self.rear+1) % self.size
        self.queue[self.rear] = val

    def dequeue(self):
        if self.front == -1: print("Empty"); return
        val = self.queue[self.front]
        if self.front == self.rear: self.front = self.rear = -1
        else: self.front = (self.front+1)%self.size
        return val
```

### **b. Priority Queue**

* Elements have priority; highest priority dequeued first
* Implemented using `heapq` (min-heap by default)

```python
import heapq

pq = []
heapq.heappush(pq, (2, "Medium"))
heapq.heappush(pq, (1, "High"))
heapq.heappush(pq, (3, "Low"))

while pq:
    print(heapq.heappop(pq)[1])
```

* Add/remove from **both ends**
* Implemented via `collections.deque`

```python
from collections import deque

dq = deque([1,2,3])
dq.append(4)      # add rear
dq.appendleft(0)  # add front
dq.pop()          # remove rear
dq.popleft()      # remove front
```

## **3. Applications**

### **a. BFS (Graph Traversal)**

```python
from collections import deque

graph = {'A':['B','C'],'B':['D'],'C':['E'],'D':[],'E':[]}
def bfs(start):
    visited = set()
    q = deque([start])
    while q:
        node = q.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            q.extend(graph[node])
bfs('A')  # Output: A B C D E
```

---

### **b. CPU Scheduling (Round Robin)**

```python
tasks = deque([{"name":"T1","burst":5},{"name":"T2","burst":3}])
time_slice = 2
while tasks:
    t = tasks.popleft()
    t['burst'] -= time_slice
    if t['burst']>0: tasks.append(t)
    else: print(f"{t['name']} completed")
```

#### **FIFO Cache**

```python
from collections import deque
cache = deque(maxlen=3)
for page in [1,2,3,4]:
    cache.append(page)
    print(list(cache))
```

#### **LRU Cache**

```python
from collections import OrderedDict
lru = OrderedDict()
size = 3
pages = [1,2,3,4,2]
for p in pages:
    if p in lru: lru.move_to_end(p)
    else:
        if len(lru)>=size: lru.popitem(last=False)
        lru[p]=True
    print(list(lru.keys()))
```

## 🧠 What is a Linked List?

A **Linked List** is a **linear data structure** where elements (called **nodes**) are **not stored in contiguous memory locations**.
Each node **stores data** and a **reference (pointer)** to the next node in the sequence.

Unlike arrays, which are indexed, linked lists must be **traversed sequentially**.

---

## ⚙️ Structure of a Node

A **node** is a fundamental building block:

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None  # pointer to the next node
```

---

## 🧩 1. Singly Linked List

Each node points only to the **next** node.
Traversal is **one direction only** (from head to tail).

### Example Diagram

```
HEAD → [10 | *] → [20 | *] → [30 | None]
```

### Implementation

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # 🟢 Insertion at end
    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # 🔴 Deletion by value
    def delete(self, key):
        temp = self.head

        # If head needs to be deleted
        if temp and temp.data == key:
            self.head = temp.next
            temp = None
            return

        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next

        if not temp:
            return  # key not found

        prev.next = temp.next
        temp = None

    # 🔁 Traversal
    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end=" → ")
            temp = temp.next
        print("None")

    # 🔁 Reversal
    def reverse(self):
        prev = None
        current = self.head
        while current:
            nxt = current.next
            current.next = prev
            prev = current
            current = nxt
        self.head = prev
```

---

## 🧩 2. Doubly Linked List

Each node points to both **previous** and **next** nodes.
It allows **bidirectional traversal**.

### Diagram

```
None ← [10 | * | *] ↔ [20 | * | *] ↔ [30 | * | None]
```

### Implementation

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_end(self, data):
        new_node = DNode(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    def delete(self, key):
        temp = self.head
        while temp and temp.data != key:
            temp = temp.next

        if not temp:
            return

        if temp.prev:
            temp.prev.next = temp.next
        else:
            self.head = temp.next  # deleting head

        if temp.next:
            temp.next.prev = temp.prev

    def traverse_forward(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ↔ ")
            last = temp
            temp = temp.next
        print("None")

    def traverse_backward(self):
        temp = self.head
        while temp.next:
            temp = temp.next
        while temp:
            print(temp.data, end=" ↔ ")
            temp = temp.prev
        print("None")
```

---

## 🧩 3. Circular Linked List

The **last node points back to the first** node — forming a circle.

### Diagram

```
HEAD → [10 | *] → [20 | *] → [30 | *] ─┐
        └───────────────────────────────┘
```

### Implementation

```python
class CNode:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_end(self, data):
        new_node = CNode(data)
        if not self.head:
            self.head = new_node
            new_node.next = new_node
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    def traverse(self):
        if not self.head:
            return
        temp = self.head
        while True:
            print(temp.data, end=" → ")
            temp = temp.next
            if temp == self.head:
                break
        print("(back to head)")
```

## 🧠 What is LRU Cache?

An **LRU Cache** stores a limited number of items. When it reaches capacity:

* The **least recently used** item is removed to make space.
* Most recently accessed items are **moved to the front** (or marked as MRU).

We need **O(1) operations** for:

1. `get(key)` → retrieve value and mark as recently used
2. `put(key, value)` → insert/update value and maintain LRU order

---

## ⚙️ Why Doubly Linked List + HashMap?

* **Doubly Linked List (DLL)**

  * Maintains **usage order**.
  * Front = Most Recently Used (MRU), Back = Least Recently Used (LRU).
  * Easy to **add/remove nodes in O(1)**.

* **HashMap (dict)**

  * Maps `key → node` for **O(1) access**.
  * Without this, searching a node in DLL would be O(n).

---

## 🔹 Implementation

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.prev = None
        self.next = None


class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = {}  # key -> Node

        # Dummy head & tail nodes to avoid edge cases
        self.head = Node(0, 0)
        self.tail = Node(0, 0)
        self.head.next = self.tail
        self.tail.prev = self.head

    # Remove node from DLL
    def _remove(self, node):
        prev_node = node.prev
        next_node = node.next
        prev_node.next = next_node
        next_node.prev = prev_node

    # Insert node right after head (mark as MRU)
    def _insert(self, node):
        node.next = self.head.next
        node.prev = self.head
        self.head.next.prev = node
        self.head.next = node

    # Get value by key
    def get(self, key):
        if key in self.cache:
            node = self.cache[key]
            self._remove(node)
            self._insert(node)  # Move to front (MRU)
            return node.value
        return -1  # Key not found

    # Put key-value pair
    def put(self, key, value):
        if key in self.cache:
            self._remove(self.cache[key])
        node = Node(key, value)
        self._insert(node)
        self.cache[key] = node

        # Check capacity
        if len(self.cache) > self.capacity:
            # Remove LRU node (node before tail)
            lru = self.tail.prev
            self._remove(lru)
            del self.cache[lru.key]
```

---

## 🔹 Example Usage

```python
cache = LRUCache(3)

cache.put(1, "A")
cache.put(2, "B")
cache.put(3, "C")

print(cache.get(1))  # A → moves 1 to front (MRU)
cache.put(4, "D")    # Evicts 2 (LRU)
print(cache.get(2))  # -1 (not found)
cache.put(5, "E")    # Evicts 3 (LRU)
print(cache.get(3))  # -1
print(cache.get(4))  # D
print(cache.get(5))  # E
```

**Output:**

```
A
-1
-1
D
E
```
