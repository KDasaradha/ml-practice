## **1️⃣ Tree Terminology**

A **tree** is a hierarchical data structure consisting of nodes. It is widely used to represent hierarchical relationships (like file systems, organizational charts, etc.).

**Key terms:**

1. **Node** – Fundamental unit of a tree. It contains data and references to its child nodes.
2. **Root** – The topmost node of the tree (no parent).
3. **Child** – Node directly connected to another node when moving away from the root.
4. **Parent** – Node with one or more children.
5. **Leaf** – Node with no children.
6. **Sibling** – Nodes sharing the same parent.
7. **Edge** – Connection between parent and child nodes.
8. **Path** – Sequence of nodes from one node to another.
9. **Height of a node** – Longest path from that node to a leaf.
10. **Depth of a node** – Number of edges from the root to that node.
11. **Height of tree** – Height of the root node.
12. **Subtree** – Tree formed by a node and all its descendants.

**Visual Example:**

```
         1        <- Root
       /   \
      2     3    <- 2 and 3 are children of 1
     / \      \
    4   5      6 <- 4,5,6 are leaves
```

* Root = 1
* Leaf nodes = 4, 5, 6
* Height of node 2 = 1 (max depth from node 2 to its leaf)
* Depth of node 5 = 2 (edges from root 1 → 2 → 5)

## **2️⃣ Tree Traversal Techniques**

Traversal is visiting all nodes **systematically**.

### **A. Depth-First Traversals (DFS)**

1. **Preorder (Root → Left → Right)**

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

# Example
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
preorder(root)  # Output: 1 2 4 5 3
```

2. **Inorder (Left → Root → Right)**

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

inorder(root)  # Output: 4 2 5 1 3
```

3. **Postorder (Left → Right → Root)**

```python
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

postorder(root)  # Output: 4 5 2 3 1
```

### **B. Breadth-First Traversal (BFS / Level Order)**

```python
from collections import deque

def level_order(root):
    if not root:
        return
    queue = deque([root])
    while queue:
        node = queue.popleft()
        print(node.val, end=" ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

level_order(root)  # Output: 1 2 3 4 5
```

## **3️⃣ Binary Tree**

A **Binary Tree** is a tree in which each node has **at most two children**: left and right.

### **A. Binary Tree Creation**

We can create a binary tree using a `Node` class:

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

# Example Tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
```

Visual representation:

```
        1
       / \
      2   3
     / \
    4   5
```

---

### **B. Binary Tree Traversal**

Traversal works exactly as we discussed earlier:

* **Preorder:** Root → Left → Right → `1 2 4 5 3`
* **Inorder:** Left → Root → Right → `4 2 5 1 3`
* **Postorder:** Left → Right → Root → `4 5 2 3 1`
* **Level order:** `1 2 3 4 5`

We already have the Python functions for this.

---

### **C. Height of Binary Tree**

**Height** = Number of edges in the longest path from root to a leaf.

```python
def height(root):
    if not root:
        return 0
    return 1 + max(height(root.left), height(root.right))

print("Height:", height(root))  # Output: 3
```

---

### **D. Diameter of Binary Tree**

**Diameter** = Length of the longest path between any two nodes (edges counted).

```python
def diameter(root):
    diameter_result = [0]

    def dfs(node):
        if not node:
            return 0
        left = dfs(node.left)
        right = dfs(node.right)
        diameter_result[0] = max(diameter_result[0], left + right)
        return 1 + max(left, right)

    dfs(root)
    return diameter_result[0]

print("Diameter:", diameter(root))  # Output: 4
```

---

### **E. Path Sum**

Path sum = sum of values along a path. Common problem: **Check if a path sum exists** from root to leaf.

```python
def has_path_sum(root, target):
    if not root:
        return False
    if not root.left and not root.right:  # Leaf node
        return root.val == target
    return has_path_sum(root.left, target - root.val) or has_path_sum(root.right, target - root.val)

print(has_path_sum(root, 7))  # True (1->2->4)
print(has_path_sum(root, 10)) # True (1->2->5)
```

## **4️⃣ Binary Search Tree (BST)**

A **Binary Search Tree** is a **binary tree** with the property:

> For each node, all values in the left subtree are **smaller**, and all values in the right subtree are **larger**.

This property allows **efficient searching, insertion, and deletion**.

---

### **A. BST Node and Creation**

```python
class BSTNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

---

### **B. BST Insertion**

```python
def insert(root, key):
    if not root:
        return BSTNode(key)
    if key < root.val:
        root.left = insert(root.left, key)
    elif key > root.val:
        root.right = insert(root.right, key)
    return root

# Example
root = None
for val in [50, 30, 70, 20, 40, 60, 80]:
    root = insert(root, val)
```

Visual BST:

```
        50
       /  \
     30    70
    / \    / \
   20 40  60 80
```

---

### **C. BST Searching**

```python
def search(root, key):
    if not root:
        return False
    if key == root.val:
        return True
    elif key < root.val:
        return search(root.left, key)
    else:
        return search(root.right, key)

print(search(root, 60))  # True
print(search(root, 25))  # False
```

---

### **D. BST Deletion**

Deletion has 3 cases:

1. **Leaf node** → Remove it directly.
2. **Node with one child** → Replace it with its child.
3. **Node with two children** → Replace with **inorder successor** (smallest in right subtree).

```python
def min_value_node(node):
    current = node
    while current.left:
        current = current.left
    return current

def delete(root, key):
    if not root:
        return root
    if key < root.val:
        root.left = delete(root.left, key)
    elif key > root.val:
        root.right = delete(root.right, key)
    else:
        if not root.left:  # Case 1 or 2
            return root.right
        elif not root.right:
            return root.left
        # Case 3: two children
        temp = min_value_node(root.right)
        root.val = temp.val
        root.right = delete(root.right, temp.val)
    return root

# Example
root = delete(root, 50)  # Delete root
```

---

### **E. Balanced vs Unbalanced BST**

1. **Balanced BST**

   * Height is approximately `log₂(n)`
   * Example: **AVL Tree, Red-Black Tree**
   * Operations (search, insert, delete) → **O(log n)**

2. **Unbalanced BST**

   * Can degenerate into a linked list (all nodes on one side)
   * Operations → **O(n)**

**Example of unbalanced BST:**

```
Insert 1,2,3,4,5 → Right-skewed tree
1
 \
  2
   \
    3
     \
      4
       \
        5
```

---

### ✅ **Summary**

* BST allows **efficient search, insertion, and deletion**.
* Handled **all deletion cases**.
* Balanced BST keeps operations fast, unbalanced can slow down to O(n).

## **5️⃣ Heaps**

A **Heap** is a **complete binary tree** satisfying the **heap property**:

1. **Min Heap** – Parent node ≤ children

   * Root contains the **smallest** element.
2. **Max Heap** – Parent node ≥ children

   * Root contains the **largest** element.

> Heaps are often used to implement **priority queues**.

---

### **A. Heap Implementation in Python**

Python provides the `heapq` module, which implements a **Min Heap** by default.

```python
import heapq

# Min Heap
min_heap = []
heapq.heappush(min_heap, 5)
heapq.heappush(min_heap, 1)
heapq.heappush(min_heap, 10)
heapq.heappush(min_heap, 3)

print("Min Heap:", min_heap)  # Internal structure: [1, 3, 10, 5]

# Extract min
print("Pop:", heapq.heappop(min_heap))  # 1
print("Heap after pop:", min_heap)      # [3, 5, 10]
```

**Max Heap** can be implemented by inserting negative values:

```python
max_heap = []
for val in [5,1,10,3]:
    heapq.heappush(max_heap, -val)

# Pop max
print("Pop Max:", -heapq.heappop(max_heap))  # 10
```

---

### **B. Heap Operations Complexity**

| Operation       | Time Complexity |
| --------------- | --------------- |
| Insert          | O(log n)        |
| Extract min/max | O(log n)        |
| Peek min/max    | O(1)            |
| Heapify a list  | O(n)            |

---

### **C. Applications of Heaps**

1. **Priority Queue**

   * Task scheduling, Dijkstra’s algorithm, A* search.

```python
import heapq

tasks = [(3, "low"), (1, "high"), (2, "medium")]
heapq.heapify(tasks)  # Converts list into a min-heap based on first element
while tasks:
    print(heapq.heappop(tasks)[1])
# Output: high, medium, low
```

2. **Top-K problems**

   * Find the **k largest** or **k smallest** elements efficiently.

```python
nums = [5,3,8,1,2,10]
k = 3
print(heapq.nlargest(k, nums))  # [10, 8, 5]
print(heapq.nsmallest(k, nums)) # [1, 2, 3]
```

---

### ✅ **Summary**

* Heaps are **complete binary trees** with **min/max property**.
* Python’s `heapq` is a **min heap**. Max heap requires negation trick.
* Applications: **priority queues**, **Top-K problems**, **efficient sorting (heap sort)**.

## **6️⃣ Trie (Prefix Tree)**

A **Trie** is a tree-like data structure used to **store strings efficiently** for **prefix-based searching**.

* Each **node** represents a character.
* **Root** is empty or null.
* Words are formed by traversing from root to leaf.
* Nodes may have a flag to mark the **end of a word**.

**Example:**

Words: `"cat"`, `"car"`, `"dog"`

```
       (root)
       /   \
      c     d
     / \     \
    a   a     o
   /     \     \
  t       r     g
(end)  (end)  (end)
```

---

### **A. Trie Node Structure**

```python
class TrieNode:
    def __init__(self):
        self.children = {}  # dict to store children nodes
        self.is_end_of_word = False
```

---

### **B. Insert Word**

```python
class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

# Example
trie = Trie()
for word in ["cat", "car", "dog"]:
    trie.insert(word)
```

---

### **C. Search Word**

```python
def search(self, word):
    node = self.root
    for char in word:
        if char not in node.children:
            return False
        node = node.children[char]
    return node.is_end_of_word

print(trie.search("cat"))  # True
print(trie.search("ca"))   # False
print(trie.search("dog"))  # True
```

---

### **D. Delete Word**

Deleting a word in a trie is **recursive**. We delete nodes only if they are **not part of another word**.

```python
def delete(self, node, word, depth=0):
    if not node:
        return False
    if depth == len(word):
        if node.is_end_of_word:
            node.is_end_of_word = False
        return len(node.children) == 0  # True if node has no children
    char = word[depth]
    if char in node.children:
        should_delete = self.delete(node.children[char], word, depth + 1)
        if should_delete:
            del node.children[char]
            return len(node.children) == 0 and not node.is_end_of_word
    return False

# Usage
trie.delete(trie.root, "car")
print(trie.search("car"))  # False
print(trie.search("cat"))  # True
```

---

### **E. Applications of Trie**

1. **Autocomplete / Search Suggestions**

   * Predict words as you type based on prefixes.
   * Example: Google search bar, code editors.

2. **Spell Checker**

   * Quickly find if a word exists.
   * Suggest corrections using nearby prefixes.

3. **IP Routing / Network**

   * Store routing tables efficiently.

4. **Dictionary / Word Games**

   * Scrabble, Boggle – fast lookup of valid words.

---

### ✅ **Summary**

* Trie is **efficient for string storage & prefix search**.
* Operations:

  * Insert: O(L)
  * Search: O(L)
  * Delete: O(L)
    *(L = length of word)*
* Applications: **autocomplete, spell check, prefix search**.

```python
# =====================================================
# TREE TERMINOLOGY & TRAVERSAL TECHNIQUES
# =====================================================

from collections import deque
import heapq

# Node class for Binary Tree / BST
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

# -------------------
# Tree Traversals
# -------------------

# Preorder: Root -> Left -> Right
def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

# Inorder: Left -> Root -> Right
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

# Postorder: Left -> Right -> Root
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

# Level order (BFS)
def level_order(root):
    if not root:
        return
    queue = deque([root])
    while queue:
        node = queue.popleft()
        print(node.val, end=" ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

# -------------------
# Binary Tree Operations
# -------------------

# Height of binary tree
def height(root):
    if not root:
        return 0
    return 1 + max(height(root.left), height(root.right))

# Diameter of binary tree
def diameter(root):
    dia = [0]
    def dfs(node):
        if not node:
            return 0
        left = dfs(node.left)
        right = dfs(node.right)
        dia[0] = max(dia[0], left + right)
        return 1 + max(left, right)
    dfs(root)
    return dia[0]

# Check if path sum exists
def has_path_sum(root, target):
    if not root:
        return False
    if not root.left and not root.right:
        return root.val == target
    return has_path_sum(root.left, target - root.val) or has_path_sum(root.right, target - root.val)

# =====================================================
# BINARY SEARCH TREE (BST)
# =====================================================

class BSTNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

# Insert in BST
def bst_insert(root, key):
    if not root:
        return BSTNode(key)
    if key < root.val:
        root.left = bst_insert(root.left, key)
    elif key > root.val:
        root.right = bst_insert(root.right, key)
    return root

# Search in BST
def bst_search(root, key):
    if not root:
        return False
    if root.val == key:
        return True
    elif key < root.val:
        return bst_search(root.left, key)
    else:
        return bst_search(root.right, key)

# Delete in BST
def min_value_node(node):
    current = node
    while current.left:
        current = current.left
    return current

def bst_delete(root, key):
    if not root:
        return root
    if key < root.val:
        root.left = bst_delete(root.left, key)
    elif key > root.val:
        root.right = bst_delete(root.right, key)
    else:
        # Node with one or no child
        if not root.left:
            return root.right
        elif not root.right:
            return root.left
        # Node with two children
        temp = min_value_node(root.right)
        root.val = temp.val
        root.right = bst_delete(root.right, temp.val)
    return root

# =====================================================
# HEAPS
# =====================================================

# Min Heap and Max Heap using heapq
min_heap = []
max_heap = []

# Insert in Min Heap
def insert_min_heap(val):
    heapq.heappush(min_heap, val)

# Extract Min
def extract_min():
    return heapq.heappop(min_heap) if min_heap else None

# Insert in Max Heap
def insert_max_heap(val):
    heapq.heappush(max_heap, -val)

# Extract Max
def extract_max():
    return -heapq.heappop(max_heap) if max_heap else None

# Top-K using heapq
def top_k_elements(arr, k, largest=True):
    if largest:
        return heapq.nlargest(k, arr)
    else:
        return heapq.nsmallest(k, arr)

# =====================================================
# TRIE (PREFIX TREE)
# =====================================================

class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    # Insert word
    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

    # Search word
    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end_of_word

    # Delete word
    def delete(self, node, word, depth=0):
        if not node:
            return False
        if depth == len(word):
            if node.is_end_of_word:
                node.is_end_of_word = False
            return len(node.children) == 0
        char = word[depth]
        if char in node.children:
            should_delete = self.delete(node.children[char], word, depth + 1)
            if should_delete:
                del node.children[char]
                return len(node.children) == 0 and not node.is_end_of_word
        return False

# =====================================================
# ================== USAGE EXAMPLES ==================
# =====================================================

if __name__ == "__main__":
    print("=== BINARY TREE ===")
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)

    print("Preorder:", end=" ")
    preorder(root)
    print("\nInorder:", end=" ")
    inorder(root)
    print("\nPostorder:", end=" ")
    postorder(root)
    print("\nLevel order:", end=" ")
    level_order(root)
    print("\nHeight:", height(root))
    print("Diameter:", diameter(root))
    print("Has path sum 7:", has_path_sum(root, 7))

    print("\n=== BST ===")
    bst_root = None
    for val in [50, 30, 70, 20, 40, 60, 80]:
        bst_root = bst_insert(bst_root, val)
    print("BST Search 60:", bst_search(bst_root, 60))
    bst_root = bst_delete(bst_root, 50)  # delete root
    print("BST Search 50 after delete:", bst_search(bst_root, 50))

    print("\n=== HEAPS ===")
    for val in [5, 1, 10, 3]:
        insert_min_heap(val)
        insert_max_heap(val)
    print("Min Heap pop:", extract_min())
    print("Max Heap pop:", extract_max())
    print("Top 2 largest:", top_k_elements([5, 1, 10, 3], 2))

    print("\n=== TRIE ===")
    trie = Trie()
    words = ["cat", "car", "dog"]
    for word in words:
        trie.insert(word)
    print("Search 'cat':", trie.search("cat"))
    trie.delete(trie.root, "car")
    print("Search 'car' after delete:", trie.search("car"))
```
