## 🧩 3. List Comprehensions, Lambda, Map, Filter, and Reduce

These tools make your Python code **shorter, cleaner, and more functional** — they’re used heavily in data manipulation and algorithm writing.

Let’s break them down one by one 👇

---

## 🧠 3.1 List Comprehensions

### 💡 What It Is

A **concise way to create lists** using a single line.

**Syntax:**

```python
[expression for item in iterable if condition]
```

### 🔹 Example 1: Basic

```python
nums = [1, 2, 3, 4, 5]
squares = [x**2 for x in nums]
print(squares)  # [1, 4, 9, 16, 25]
```

### 🔹 Example 2: With Condition

```python
evens = [x for x in nums if x % 2 == 0]
print(evens)  # [2, 4]
```

### 🔹 Example 3: String Manipulation

```python
names = ["rami", "reddy", "python"]
upper_names = [name.capitalize() for name in names]
print(upper_names)  # ['Rami', 'Reddy', 'Python']
```

💡 **Faster and cleaner** than traditional `for` loops.

---

## ⚡ 3.2 Lambda Functions (Anonymous Functions)

A **lambda function** is a small, one-line function without a name.

**Syntax:**

```python
lambda arguments: expression
```

### 🔹 Example 1:

```python
square = lambda x: x**2
print(square(5))  # 25
```

### 🔹 Example 2:

```python
add = lambda a, b: a + b
print(add(3, 4))  # 7
```

💡 Used with `map()`, `filter()`, and `reduce()` for quick operations.

---

## 🧮 3.3 `map()` — Apply a Function to Each Item

**Syntax:**

```python
map(function, iterable)
```

Returns a **map object** → convert to list with `list()`.

### 🔹 Example:

```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)  # [1, 4, 9, 16]
```

### 🔹 Real Example:

```python
names = ["rami", "reddy"]
cap_names = list(map(str.title, names))
print(cap_names)  # ['Rami', 'Reddy']
```

---

## 🧹 3.4 `filter()` — Keep Only Matching Items

**Syntax:**

```python
filter(function, iterable)
```

The function must **return True/False**.

### 🔹 Example:

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # [2, 4, 6]
```

---

## 🧾 3.5 `reduce()` — Reduce a List to a Single Value

`reduce()` applies a function cumulatively to items.
(You must import it from `functools`.)

**Syntax:**

```python
from functools import reduce
reduce(function, iterable)
```

### 🔹 Example 1: Sum of List

```python
from functools import reduce

nums = [1, 2, 3, 4]
total = reduce(lambda a, b: a + b, nums)
print(total)  # 10
```

### 🔹 Example 2: Find Maximum

```python
max_num = reduce(lambda a, b: a if a > b else b, nums)
print(max_num)  # 4
```

# ⚙️ 1. Time and Space Complexity — The Core Idea

Every algorithm has two main “costs”:

1. **Time Complexity** — how much *time* it takes as input grows
2. **Space Complexity** — how much *memory* it uses as input grows

We measure both in **terms of input size** (usually denoted as `n`).

Example:
If your algorithm loops over a list of `n` items once → it takes **O(n)** time.

---

## ⏱️ 2. Time Complexity

Time complexity is about **how the runtime scales** with input size.

We use **asymptotic notation** to represent this growth rate — that’s where **Big O** comes in.

---

# 🧮 3. Big O Notation

Big O represents the **upper bound** (worst-case) growth rate of an algorithm’s time or space usage.

It tells us *how fast* runtime increases as input size increases.

---

## ⚡ Common Big O Complexities (with Examples)

| Big O          | Name         | Example Code                            | Description                                |
| -------------- | ------------ | --------------------------------------- | ------------------------------------------ |
| **O(1)**       | Constant     | Accessing an element in a list by index | Time doesn’t depend on input size          |
| **O(log n)**   | Logarithmic  | Binary search                           | Input size halves each step                |
| **O(n)**       | Linear       | Looping through an array once           | Grows linearly with input                  |
| **O(n log n)** | Linearithmic | Merge sort, Quick sort (avg)            | Divide and conquer pattern                 |
| **O(n²)**      | Quadratic    | Nested loops                            | For each item, loop through all others     |
| **O(2ⁿ)**      | Exponential  | Recursive Fibonacci                     | Grows very fast — doubles each n           |
| **O(n!)**      | Factorial    | Generating permutations                 | Even worse — grows faster than exponential |

---

### 💡 Examples

#### O(1)

```python
arr = [10, 20, 30]
print(arr[0])  # constant time access
```

#### O(n)

```python
for item in arr:
    print(item)
```

#### O(n²)

```python
for i in arr:
    for j in arr:
        print(i, j)
```

#### O(log n)

```python
# Binary search halves the input each step
low, high = 0, len(arr) - 1
while low <= high:
    mid = (low + high) // 2
    low = mid + 1  # or high = mid - 1
```

## 📊 4. Big Ω (Omega) and Big Θ (Theta)

| Notation    | Meaning   | Represents                 |
| ----------- | --------- | -------------------------- |
| **O(f(n))** | Big O     | Upper bound (worst case)   |
| **Ω(f(n))** | Big Omega | Lower bound (best case)    |
| **Θ(f(n))** | Big Theta | Tight bound (average case) |

Example (Linear Search in Array):

| Case        | Meaning                    | Complexity |
| ----------- | -------------------------- | ---------- |
| **Best**    | Target is first element    | Ω(1)       |
| **Worst**   | Target not found / last    | O(n)       |
| **Average** | Target somewhere in middle | Θ(n)       |

## 🧠 5. Space Complexity

Space complexity measures how much **extra memory** an algorithm needs — not including the input itself.

Example:

```python
def sum_arr(arr):
    total = 0      # one variable
    for num in arr:
        total += num
    return total
```

✅ Space Complexity = **O(1)** (constant space)
(because only one variable `total` is used regardless of array size)

Another example:

```python
def copy_list(arr):
    new_arr = arr[:]   # creates another list of size n
    return new_arr
```

🚨 Space Complexity = **O(n)** (needs memory proportional to input)

## ⚖️ 6. Amortized Analysis

Used when an algorithm occasionally performs an **expensive operation**, but that operation doesn’t happen every time.

So, instead of worst-case per operation, we analyze the **average cost over many operations**.

### Example: Dynamic Array (Python List)

Python lists automatically resize when they run out of space:

* Most `append()` calls → O(1)
* Occasionally → O(n) when list expands (copies all items)

But over time, these O(n) resizes are rare.
👉 So the **amortized time complexity** for `.append()` is **O(1)** average.

---

### 💡 Intuition for Amortized Complexity

| Operation                 | Worst Case           | Amortized Average |
| ------------------------- | -------------------- | ----------------- |
| `list.append()`           | O(n) (when resizing) | **O(1)**          |
| `push()` in dynamic stack | O(n) occasionally    | **O(1)** average  |
| Hash table insert         | O(n) when rehashing  | **O(1)** average  |
