### 1. **Basics**

* What is Seaborn & how it relates to Matplotlib.
* Importing:

  ```python
  import seaborn as sns
  import matplotlib.pyplot as plt
  ```
* Built-in datasets (`sns.load_dataset("tips")`).
* Integration with Pandas DataFrames (column names as arguments).

---

### 2. **Styling & Themes**

* Seaborn default styles vs Matplotlib.
* Setting styles: `sns.set_style("whitegrid")`.
* Context scaling: `sns.set_context("talk")`.
* Palettes: `sns.color_palette()`, built-in color schemes.

---

### 3. **Relational Plots**

* `sns.scatterplot()` â€“ scatter plots.
* `sns.lineplot()` â€“ line plots with confidence intervals.
* `sns.relplot()` â€“ high-level interface (scatter/line).

---

### 4. **Categorical Plots**

* `sns.barplot()` â€“ bar charts (with confidence intervals).
* `sns.countplot()` â€“ counts of categories.
* `sns.boxplot()` â€“ distribution + outliers.
* `sns.violinplot()` â€“ KDE + box plot.
* `sns.stripplot()` â€“ individual points.
* `sns.swarmplot()` â€“ non-overlapping points.
* `sns.catplot()` â€“ high-level interface (multiple categorical plots).

---

### 5. **Distribution Plots**

* `sns.histplot()` â€“ histograms (replacement for old `distplot`).
* `sns.kdeplot()` â€“ kernel density estimate.
* `sns.ecdfplot()` â€“ empirical cumulative distribution.
* `sns.displot()` â€“ high-level for hist/kde/ECDF.

---

### 6. **Matrix & Heatmap Plots**

* `sns.heatmap()` â€“ correlation matrices, annotations.
* `sns.clustermap()` â€“ hierarchical clustering heatmap.

---

### 7. **Regression Plots**

* `sns.regplot()` â€“ regression + scatter.
* `sns.lmplot()` â€“ high-level regression plotting.

---

### 8. **Pairwise & Multivariate Plots**

* `sns.pairplot()` â€“ scatterplot matrix with histograms.
* `sns.jointplot()` â€“ bivariate distributions with marginal plots.

---

### 9. **Facet Grids & Multi-plotting**

* `sns.FacetGrid()` â€“ multiple subplots by category.
* `sns.relplot()` / `sns.catplot()` (these use FacetGrid internally).
* Customizing FacetGrid with `map()`, `map_dataframe()`.

---

### 10. **Customization**

* Adding titles & labels (`plt.title()`, `ax.set_title()`).
* Legends & hue/size/style encoding.
* Custom palettes (`sns.set_palette()`).
* Controlling axis limits and scales.

---

### 11. **Integration with Matplotlib**

* Accessing underlying Matplotlib objects (`ax = sns.scatterplot(...); ax.set_title("Example")`).
* Mixing Seaborn plots with Matplotlib annotations.
* Fine-tuning layouts (`plt.subplots_adjust()`, `plt.tight_layout()`).
